## Feature Engineering:

### Creating columns based on popularity:

```python
df['very_highly_popular'] = df['popularity'].apply(lambda x: x>75)
df['highly_popular'] = df['popularity'].apply(lambda x: x > 50 and x <= 75)
df['less_popular'] = df['popularity'].apply(lambda x: x <=50 and x > 25)
df['very_less_popular'] = df['popularity'].apply(lambda x: x < 25)
df.head()
```
